### Question 3 (20 points) Use matplotlib to produce a plot of the function f(x) = e−x/p sin(πx) over the interval [0, 10], for the values of the parameter p = 1, 2, 5. Include labels for the x- and y-axes, and a legend explaining which line is which plot.
To plot a function, first create an array of x values, for example by using “arrange” function. Make sure that you use enough points so that the curve appears smooth. Use “plot” function to plot f(x) versus x. The shape of your plot will look like the one below (the name of the plot, legend and labels are missing).

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0,10,100)
y1 = np.exp(-x/1) * np.sin(np.pi * x) 
y2 = np.exp(-x/2) * np.sin(np.pi * x) 
y3 = np.exp(-x/3) * np.sin(np.pi * x) 

plt.plot(x, y1, color='blue', label="p = 1")
plt.plot(x, y2, color='green', label="p = 2")
plt.plot(x, y3, color='red', label="p = 3")
plt.ylim=(-0.8, 1.0)

plt.title("Plot of the function f(x)= e^(-x/p) * sin(pi*x)")
plt.legend(loc= 'best')
plt.xlabel("X")
plt.ylabel("Y = f(x)")
plt.show()

```

### Question 4 (20 points) Use matplotlib to plot the shape of a limacon for r0 = 0.8, r0 = 1.0, and r0 = 1.2. Be sure to use enough points that the curve appears smooth. Use a legend to identify which curve is which.
The shape of a limacon can be defined parametrically as:
r = r0 + cos α
x = r cos α
y = r sin α
When r0 = 1, this curve is called a cardioid.
Your plot will be similar to the one below (the name of the plot, legend and labels are missing.

```python
import numpy as np
import matplotlib.pyplot as plt

# angle 0 to 90, 50 points
alpha = np.linspace(0, 90, 360)
alpha

r0s = [(0.8, 'blue'), (1.0, 'red'), (1.2, 'green')]
for r0,color in r0s:
    r = r0 + np.cos(alpha)
    x = r * np.cos(alpha)
    y = r * np.sin(alpha)
    plt.plot(x, y, color=color, label=str(r0))

plt.title("Limacon Plots, r0 = [0.8, 1.0, 1.2]")
plt.legend(loc= 'best')
plt.xlabel("X")
plt.ylabel("Y")
plt.xlim(-0.5, 2.5)
plt.show()
```

### Question 5 (20 points) Use matplotlib to draw the trajectories of a projectile and determine the angle θmax for which the projectile will reach to longest distance on a horizontal surface. Assume initial speed V = 100 m/s and draw trajectories for θ = (0, π/2) in steps of π/8.
* Be sure to use enough points that the curves appear smooth. Use a legend to identify which curve is which.

A projectile is an object moving only under the force of gravity. In modeling the motion of a projectile we assume no air resistance. In this case, the projectile follows a parabolic path. If an object is projected at an angle θ to the ground with an initial speed V, then the path of the projectile is given by the parametric equations:
x = V t cos(θ)
y = V t sin(θ) – (gt2)/2
t represents time (t = 0 when the object is projected) and g is the acceleration due to gravity.

The value of g depends upon which system of units we use to measure distance and time. In the metric system we use meters and seconds. In this case an approximation to g is 9.8 meters/second/second (9.8 m/s/s). The initial speed is measured in meters per second.

Your plot will be similar to the one below. Note that some of the curves, the name of the plot, legend and labels are missing.

```python
import numpy as np
import matplotlib.pyplot as plt

# angle 0 to 90, 50 points
t = np.arange(0, 1200, 1)
V = 100 
g = 9.8

thetas = [(0, 'black'),(np.pi/8, 'blue'), (np.pi/6, 'red'), 
          (np.pi/4, 'green'), (np.pi/2, 'purple')]
for theta,color in thetas:
    x = V * t * np.cos(theta)
    y = V * t * np.sin(theta) - ((g * t**2)/2)
    plt.plot(x, y, color=color, label=str(theta))

plt.title("Trajectory Plots")
plt.legend(loc= 'best')
plt.xlabel("Distance in Meters")
plt.ylabel("Altitude in Meters")
plt.xlim(0, 1200)
plt.ylim(-50, 400)
plt.show()
```
