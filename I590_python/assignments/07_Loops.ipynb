### Question 1 (25 points) Write a program that prompts the user to enter a sequence of numbers and calculates and prints:
* Total count of entered numbers
* The smallest of entered numbers
* The largest of entered numbers
* The count of positive numbers
* The count of negative numbers

The program terminates when user enters an empty string (hits Enter).
The program has to give a meaningful output in case the user enters 0, 1 or more numbers.

````python
def main():
    # prompt user to enter sequence of numbers and calculate and print:
    numStr = input("Enter a sequence of numbers: \n") 
    nums = list(map(int, numStr.split(",")))
    nums.sort(key=int)
    print(nums)

    # Total count of entered numbers
    n = len(nums)
    print("The total count of numbers is", n)
    
    # The smallest of entered numbers
    print("The smallest number in the list is", nums[0])    
    
    # The largest of entered numbers
    print("The largeest number in the list is", nums[-1])   

    # Count the positive and negative numbers
    posint = []
    negint = []
    for x in nums:
        if (x >= 0):
            posint.append(x)
        if (x < 0):
            negint.append(x)
    print("There are", len(posint), "positive numbers.")  
 
    # count of negative numbers
    print("There are", len(negint), "negative numbers.")  
    
main()
````

Nested lists are often used to represent matrices, explanation is available in:
http://openbookproject.net/thinkcs/python/english3e/lists.html

For example, the matrix: | 1  2 |
                         | 3  4 | 
                         | 5  6 |
                         
may be represented as a list with three elements, where each element is a row of the matrix:
```
mx = [[1, 2], [3, 4], [5, 6]]
```
We can select an entire row from the matrix by slicing:
```
mx[1]
[3, 4]
```
We can extract a single element from the matrix using the double-index form:
```
mx[1][2]
4
```
The first index selects the row, and the second index selects the column.
Matrix transpose ( see https://en.wikipedia.org/wiki/Transpose) is an operator which flips a matrix A over its diagonal, that is it switches the row and column indices of the matrix by producing another matrix denoted as AT. For example, if below left is a matrix A, than below right is AT.
```
    mx = | 1  2 |       mx^T = | 1 3 5 |
         | 3  4 |              | 2 4 6 |   
         | 5  6 | 
```
Formally, the i th row, j th column element of AT is the j th row, i th column element of A. NOTE: if A is an m × n matrix then A^T is an n × m matrix.

````python
# Assignment 7-question2.py Solution
# This is a program to print matrix mx and tranporsed matrix tx

def inData (m, n):
    # two nested loops: outer loop appends rows to mx
    mx = []
    for i in range(m):
        row = []
        # inner loop creates lists that represents rows
        for j in range(n):
            j = input("Enter element 'j':")
            row.append(j)
        mx.append(i)
    return

def transMx(mx):
    # double nested loops: outer loop appends rows to mx
    tx = [ ]
    for i in mx:
        row = []
    # inner loop creates lists that represents rows.
        for j in row:
            element = mx[j, i]
            tx.append(element(j))
            tx.append(i)
    return

def printMx(mx):
    # Slices and prints each row of mx in a separate line. 
    print("First row of tx is: ", m[1])
    # use syntax m[i] to extract element i from mx,
    print("Second row of tx is: ", m[2])
    # a list that contains the elements of the row i.
    print("Third row of tx is: ", m[3])
    
def main():
    # Prompt user for the dimensions m and n of a matrix.
    m = int(input("Enter number of rows, m: "))
    n = int(input("Enter number of columns, n: "))
    # calls inData (m, n) to get a matrix mx from the user.
    inData(m, n)
    # Calls printMx(mx) to print matrix mx row by row.
    printMx(mx)
    # Call transMx(mx) to get transposed matrix tx = mxT
    transMx(mx)
    # Calls printMx(tx) to print matrix tx row by row.
    printMx(tx)

main()
````

### Question 3 (25 points) Write a function avgFile(inFile) that reads a file that contains numbers separated by spaces, calculates and prints the average of numbers in each line.
Function main() prompts the user for the name of the input file and calls avgFile(inFile) to calculate and print the average of numbers in each line

Function avgFile(inFile)
Input is the file object inFile, which corresponds to the file which is open for reading.
Returns none.
Algorithm: Reads a line, calculates and prints the average of the numbers in the line with precision of 2 decimal points. Continues the same until all lines from the file are processed.

Function main():
* Prompts the user for the name of the input file
* Opens the file
* Calls the function avgFile(inFile) with the file object which corresponds to the file
* Closes the file

```
Example:
main()
What file are the numbers in? num1.txt
Average in line number 1 is 20.00
Average in line number 2 is 26.33
Average in line number 3 is 41.00
Average in line number 4 is 19.86

```
NOTE: Lecture 4.4 describes the formatting of output strings. For example, if you want to print integer lineNumber and a decimal number average with 2 decimal points you can use the statement:
```
print("Average in line number {0} is {1:0.2f}".format(lineNumber, average))
```

````python
def avgFile(inFile): 						       # Returns none
    sum = 0.0
    count = 0		
    # Reads line 
    line = inFile.readline()
    
    while line != " ":                         # if empty string, exit outer loop    
        # Calculates average of numbers in line
        for x in map(int, line.split(",")):    # Inner loop
            sum = sum + x	 	  
            count = count + 1                  # update sum, count in inner loop
            average = sum / count        
        line = inFile.readline()               # read next line in file
    
    # Print average of numbers in line with 2 decimal points.     
    print("Average in line number {0} is {1:0.2f}".format(line, average))
    
    # Continues until all lines from file are processed. 				      

def main():
    # prompt user for name of input file
    fileName = input("What file are the numbers in?" )
    # Opens the file
    inFile = open(fileName, 'r')
    # Calls avgFile(inFile) with file object which corresponds to file
    avgFile(inFile)
    # Closes the file
    fileName.close()

main() 
````

````python
# Question3.py SOLUTION
#   Print average value of input file
#   Date: 02/21/2017   By: TA's of I590

def avgFile(inFile):
    line = inFile.readline()
    avgList = []

    while line != "":
        # update sum and count for values in line
        sumLine = 0
        countLine = 0

        for xStr in line.split(" "):
            sumLine += float(xStr)
            countLine += 1

        avgList.append(sumLine / countLine)
        line = inFile.readline()
    return avgList


def main():
    fileName = input("What file are the numbers in? ")
    inFile = open(fileName, 'r')
    avgList = avgFile(inFile)

    for i in range(len(avgList)):
        print("Average in line number {0} is {1:0.2f}".format(i+1, avgList[i]))

main()
````
