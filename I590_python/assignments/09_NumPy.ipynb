### Write a Python program that uses NumPy to calculate, print and save in a file:
* The final scores (as numbers) and corresponding letter grades of students. 
* The letter grades are explained in Assignment 6, question 1.

All the homework and lab assignments, final project and midterms are graded with maximum of 100 points. The quizzes are graded with maximum of 10 points.

Note that the assignments in the same grade group may have different weights. For example, here are the weights for the homework assignments:
```
    Homework     Weight
    HW0           50
    HW1          100
    HW2          100
    HW3          100
    HW4           50
    HW5          100
```
This means that HW0 and HW4 contribute to the homework grade by 50% compared to the other homework that contribute by 100%. The maximum number of points for homework is 500 points. For example, the table below shows the homework weights and scores for the first student in the file.
```
          HW0  HW1  HW2  HW2  HW3  HW4
Weights:   50  100  100  100   50  100
Scores:    90  100   92   90   40  100
```
You have to scale the scores using the weights to get the actual points for each homework. In the above example, the student will have 447 points for the homework grade group (out of maximum 500 points).

#### HINTS:

One way to solve this problem is to use a “for” loop and in iteration i process the scores of student in the row i. An algorithms could look like this:

You may want to use a list to store letter grades, because NumPy will most likely complain about mixing numbers and strings in a same array. If you convert the list with letter grades to an array, than you will be able to use “np.hstack” function to merge two arrays into the final array with id numbers, final scores and letter grades.

````python
# Grades2_numpy.py
import numpy as np

def letterGrade(score):    
    # Uses multi-way decision, converts the score into a letter grade.
    if (93 <= score) and (score <= 100):
        return 'A!'
    elif (90 <= score) and (score <= 92.99):
        return 'A-'
    elif (86 <= score) and (score <= 89.99): 		
        return 'B+'
    elif (83 <= score) and (score <= 85.99): 		
        return 'B'
    elif (80 <= score) and (score <= 82.99): 		
        return 'B-'
    elif (76 <= score) and (score <= 79.99):
        return 'C+'
    elif (73 <= score) and (score <= 75.99):
        return 'C'
    elif (70 <= score) and (score <= 72.99):
        return 'C-'
    elif (66 <= score) and (score <= 69.99): 		
        return 'D+'
    elif (60 <= score) and (score <= 65.99):
        return 'D'
    else:
        return 'F'

def main():
    # load the data
    scores = np.load("Grades_2.npy")
    # Change missing score to 0, data[19,17]=0; extract id as separate array
    scores[scores<0] = 0
    row, col = scores.shape
    hw_avg = []
    labs_avg = []
    quiz_avg = []
    tot_avg = []
    letters = []
    # Use “for” loop and iteration i process scores of student in row i. 
    for i in range(1, row):    
        # calculate weighted group scores and final score for each student
        # Six Homework: columns [1:7]
        # Apply weighting to HW score columns: HW0, HW4
        scores[i,1] = 0.5 * scores[i,1] 
        scores[i,5] = 0.5 * scores[i,5]
        hw = scores[i, 1:7]
        hw_avg = np.sum(hw)/5
        hw_avg = np.array(hw_avg)    
        
        # Lab scores to be weighted: Labs 1, 2, 3 - columns 7, 8, 9 
        scores[i,7] = 0.5 * scores[i,7] 
        scores[i,8] = 0.5 * scores[i,8]
        scores[i,9] = 0.5 * scores[i,9]
        labs = scores[i, 7:18]
        labs_avg = np.sum(labs)/9.5
        labs_avg = np.array(labs_avg)  
        
        # Quiz score to be weighted: Quiz 1, columns 
        scores[i,20] = 0.5 * scores[i,20] 
        quiz = scores[i, 20:29]
        quiz_avg = np.sum(quiz)/0.85        
        quiz_avg = np.array(quiz_avg)          
        
        # exam scores and final project
        mid1 = scores[i, 19].astype(np.float64)
        mid2 = scores[i, 29].astype(np.float64)
        proj = scores[i, 18].astype(np.float64)      

        # Calculate final average score by stacking grade points, take average
        total = np.hstack((hw_avg, labs_avg, quiz_avg, mid1, mid2, proj))
        tot_avg = np.mean(total)
        tot_avg = np.array(np.round(tot_avg, 2))
 
        # Call function letterGrade() to convert final scores into letter grade
        # Cast list of letter grades to array
        letters = np.asarray(letterGrade(tot_avg))
        id = np.array((scores[i,:1]))
        
        # Use np.hstack to combine id, final score, letter grade, print and save
        grades2 = np.hstack((id, tot_avg, letters))
        print(grades2)
        np.save("letter_grades2", grades2)

main()
````
