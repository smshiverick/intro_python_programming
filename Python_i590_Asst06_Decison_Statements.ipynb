### Question 1 (25 points)
Your letter grade in the Python course will be determined from your total score:
``` 
Total Score Letter Grade
93% - 100% A
90% - 92.99% A-
86% - 89.99% B+
83% - 85.99% B
80% - 82.99% B-
76% - 79.99% C+
73% - 75.99% C
70% - 72.99% C-
66% - 69.99% D+
60% - 65.99% D
0% - 59.99% F
```
### Write a function letterGrade(score) which receives a score and uses decision structure to calculate and return the corresponding letter grade.
Function letterGrade(score)
* Input score is a positive floating point number between 0 and 100.
* Returns letter grade, for example A- or B+.
* Algorithm: Uses a multi-way decision and the above table to convert the score into a letter grade.

Function main()
* No Input and no output
* Algorithm:
    * Prompts the user for your score, a positive floating point number between 0 and 100
    * Calls the function letterGrade(score)
    * Prints the letter grade received from the function letterGrade(score)

```        
Example:
Enter your total score: 94.3
Your letter grade is A!
```

````python
# letterGrade.py
# Input score is positive float between 0 and 100.
# Returns letter grade, for example A- or B+.

def letterGrade(score):    
    # Uses multi-way decision, converts the score into a letter grade.
    if 93 <= score <= 100:
        return 'A!'
    elif 90 <= score <= 92.99:
        return 'A-'
    elif 86 <= score <= 89.99: 		
        return 'B+'
    elif 83 <= score <= 85.99: 		
        return 'B'
    elif 80 <= score <= 82.99: 		
        return 'B-'
    elif 76 <= score <= 79.99:
        return 'C+'
    elif 73 <= score <= 75.99:
        return 'C'
    elif 70 <= score <= 72.99:
        return 'C-'
    elif 66 <= score <= 69.99: 		
        return 'D+'
    elif 60 <= score <= 65.99:
        return 'D'
    else:
        return 'F'

def main():
    # Promptuser for your score, a positive floating point number between 0 and 100
    score = float(input("Enter your total score: ")) 
    # Calls the function letterGrade(score)
    letter = letterGrade(score)
    # Prints letter grade from function letterGrade(score)
    print("Your letter grade is:", letter)

main()
````

### Question 2 (25 points)
The body mass index (BMI) is a statistic developed by Adolphe Quetelet in the 1900’s for evaluating body mass. It uses the same formula for men as for women and children.

According to http://www.bmi3d.com/formula.html the body mass index is calculated as kilograms divided by height in meters squared:
#### BMI = kg / m^2

Where:
* x = body weight in kilograms
* y = height in meters
* BMI in range 19 to 25, inclusive, is considered healthy.

### Write a function bodyMassIndex(height, weight) which receives height in inches and weight in pounds, converts height to meters and weight to kilograms, then calculates and returns BMI.

Function main() prompts for person’s height and weight, calls bodyMassIndex(height, weight) to determine BMI and prints the BMI and a message telling whether the person is above, within or below the healthy range.

```
Example:
Enter you height in inches: 70
Enter your weight in pounds: 170
Your BMI is 24.41, you are in the healthy range!
```

````python
def bodyMassIndex(height, weight): 
    # convert height in inches to meters 
    meters = height * 0.0254 
    # convert weight in pounds to kilograms
    kilos = weight * 0.454
    # calculates BMI as BMI kilograms divided by meters squared
    BMI = kilos / meters**2
    return BMI

def main():
    # prompts for person’s height and weight, 
    height = eval(input("Enter your height in inches: ")) 
    weight = eval(input("Enter your weight in pounds: ")) 
    # call bodyMassIndex(height, weight) to determine BMI 
    bmi = bodyMassIndex(height, weight)
    # prints message whether person is above, within or below the healthy range.
    if 19 <= bmi <= 25:
        print("Your BMI is", format(round(bmi,2)),", you are in the healthy range!")
    elif bmi < 19: 
        print("Your BMI is", format(round(bmi,2)),", you are below the healthy range.")
    else: 
        print("Your BMI is", format(round(bmi,2)),", you are above the healthy range.")

main()
````

### Question 3 (25 points) Write the following functions:
* inDate() that prompts user for a month, day and year and returns a list [month, day, year].
* isLeapYear(year) that returns True if the year is a leap year, and False otherwise.
* isMonth(month) that returns True if the month is a correct month and False otherwise.
* isDay(day) that returns True if the day is a correct day, and False otherwise.
* main() that uses function inDate() to get a date as a list [month, day, year] and calls the other three functions to determine if the [month, day, year] is a correct date in a leap

The structure of the program Question3.py is given in the following diagram.
```
                | main |
| inDate | isLeapYear | isMonth | isDay |
```

Function inDate ()
* No input.
* Returns list [month, day, year]
* Algorithm: Prompts the user for month, day, and year, creates and returns a list [month, day, year]. For example: [“February”, 12, 2017].

Function isLeapYear(year)
* Input year is a positive integer.
* Returns True if the year is a leap year, and False otherwise
* Algorithm: A year is a leap year if it is divisible by 4, unless it is a century year that is not divisible by 400. For example, 2000 is a leap year, but 1900 is not.

Function isMonth(month)
* Input month is a string.
* Returns True if the month is a correct month, and False otherwise
* Algorithm: String “month” should represent a correct month name.

Function isDay(month, day)
* Input month is a string which represents a month, day is a positive integer which represents day in a month.
* Returns True if the day is a correct day, and False otherwise
* Algorithm: day is correct if it is in the range of the days in the month. For example, 31 is a correct day in month March, but not in the month April.

Function main()
* No Input and no output
* Algorithm: Calls inDate () to obtain a date as a list [month, day, year].
    * Calls isMonth(month), prints a message “Incorrect month” if the month is incorrect and exits (statement return).
    * Calls isDay(day), prints a message “Incorrect day” if the day is incorrect and exits (statement return).
    * Checks year and prints a message “Incorrect year” if year is not a positive integer between 1 and 2017, inclusive, and exits (statement return).
    * Calls isLeapYear(year), prints a message “year is not a leap year” if the year is not a leap year, and exits (statement return).
    * Prints message “You input a correct date in a leap year” if all previous checks are passed.

````python
def inDate(): 
    # split date [month, day, year] and returns a list: [“February”, 12, 2017]
    month = input("Enter a month: ")
    day = input("Enter a day: ")
    year = input("Enter a year: ")

def isLeapYear(year):
    # use mutliway decision to identify if year is leap year
    # leap year is divisible by 4 or century year not divisible by 400. 
    if year % 4 == 0: 
        return True  
        if year % 100 == 0:
            if year % 400 != 0:
                return True
            else:
                return False
    # if year is not leap year, return False
    else:
        return False

def isMonth(month): 
    # Check that string “month” represent a correct month name.
    months = ["January", "February", "March", "April", "May", "June", "July", "August", "September", "October", "November", "December"]
    # returns True if the month is a correct month 
    if month in months:
        return True
    # returns False otherwise.
    else:
        return False
        
def isDay(month, day):
    months = ["January", "February", "March", "April", "May", "June", "July", "August", "September", "October", "November", "December"]
    index30 = [3, 5, 8, 10]
    index31 = [0, 2, 4, 6, 7, 9, 11]
    # returns True if days is in range of the number of days in the month 
    if month == months[1] and (day <= 29):
        return True
    elif month ==[months[i] for i in index30] and (day <= 30):
        return True
    elif month ==[months[i] for i in index31] and (day <= 31):
        return True
    # returns False otherwise.
    else:
        return False

def main():
    # Calls inDate () to obtain a date as a list [month, day, year].
    date = inDate()
    month = date[0]
    day = date[1]
    year = date[2]
    # Calls isMonth(month), prints message “Incorrect month” if month is incorrect 
    # and exits (statement return).
    monthCheck = isMonth(month)
    if monthCheck == False:
        print("Incorrect month")
    return
    # Calls isDay(day), prints a message “Incorrect day” if the day is incorrect 
    # and exits (statement return).
    dayCheck = isDay(month, day)
    if dayCheck == False:
        print("Incorrect day")
    return
    # Checks year and prints a message “Incorrect year” if year is not a 
    # positive integer between 1 and 2017, inclusive, and exits (statement return).
    if (1 > year <= 2017):
        print("Incorrect year")
    return
    # Calls isLeapYear(year), prints a message “year is not a leap year” 
    # if the year is not a leap year, and exits (statement return).
    leapCheck = isLeapYear(year)
    if leapCheck == False:
        print("Year is not a leap year.")
    # Prints “You input a correct date in a leap year” if previous checks are passed.
    else:
        print("You input a correct date in a leap year")

main()
````

### Question 4 (25 points) Write a program that will keep track of the score in the game and print a player who wins the game.
Assume a sport in which two players (A and B) score points and the winner is:
• A player who has at least 3 points and wins by at least 2 points or
• A player who first scores 7 points.

#### Write a function point() that prompts the user for a player who wins a point and returns 1 if player A wins and 2 if player B wins.
#### Write function game() that keeps the score in the game and returns 1 if player A wins the game and 2 if player B wins the game.
#### Write function main() that calls the previous functions and prints the winner of the game.

The structure of the program Question4.py is given in the following diagram.
```
| main |
| game |
| point |
```

Function point()
* No input.
* Returns 1 if player A wins and 2 if player B wins.
* Algorithm:
    * Prompts the user for the player who wins a point.
    * Checks input and prints a message in case of invalid input.
    * Returns 1 if player A wins and 2 if player B wins the point.

Function game()
* No input.
* Returns 1 if player A wins the game and 2 if player B wins the game.
* Algorithm:
    * Initializes the scores of both players to 0.
    * Calls point() to determine the player who wins a point
    * Increments the score of the player who wins a point.
    * Prints the scores of both players.
    * Game is over and the winner is:
        * Player who has at least 3 points and wins by at least 2 points o
        * A player who first scores 7 points.
    * Returns 1 if player A wins and 2 if player B wins the point

Function main()
* No Input and no output
* Algorithm:
    * Calls game() to determine the winner of the game.
    * Prints “The winner is player A” or “The winner is player B.”

#### HINT: You can use for loop in function game and terminate the loop if the condition for a winner evaluates True. See the hint in Assignment 5, Question 4.

````python
def point():
    # prompt user for the player who won the point.
    point = str(input("Enter the player who won a point (A or B): "))
    # check input and print a message in case of invalid input.
    # returns 1 if player A wins and 2 if player B wins the point.
    if point == "A":
        return 1
    elif point == "B":
        return 2
    else:
        print("Invalid input, try again.")

def game():
    # Initializes scores of both players to 0.
    A = 0
    B = 0
    win = 0 
    # iterates through range of points
    for i in range(14):
        # Calls point() and increments score of player who wins point.
        score = point()
        if score == 1:
            A = A + 1
        if score == 2:
            B = B + 1
        # print("Player A score:", A,"\nPlayer B score:", B)
        # Game is over if player has at least 4  points and wins by at least 2 points 
        # or a player scores 7 points first.
        if A == 7:
            win = 1
            break
        if B == 7: 
            win = 2
            break
        if A >= 3 and (abs(A-B)) >= 2:
            win = 1
        if B >= 3 and (abs(A-B)) >= 2:
            win = 2        
    if win == 1:
        return 1
    else: 
        return 2

def main():
    # Calls game() to determine the winner of the game.
    game()
    win = game()
    # Prints “The winner is player A” or “The winner is player B.”
    if win == 1:
        print("The winner is player A")
    else: 
        print("The winner is player B")

main()
````

Sometimes, you must terminate your loop abruptly. In that case, we use “break” statement. 
It terminates the current loop and resumes execution at the next statement, just 
like the traditional break statement in C, Java, C++.

Most common use for break is when some external condition is triggered requiring 
a hasty exit from a loop. The break statement can be used in both while and for loops.
If you are using nested loops, the break statement stops the execution of the 
innermost loop and start executing the next line of code after the block.
